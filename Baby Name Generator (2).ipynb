```python
import tkinter as tk
from tkinter import *
from PIL import Image, ImageTk
```

```python
window_main=tk.Tk(className='Baby Name Generator')
window_main.geometry('600x600')

babynamesDB = ["Abdul" , "Anil", "Anand", "Aravind","Bhuvan", "Bharath", "Nick","Cindrella","Danny","Dinesh","Dravid","Elly","Frank","Ganesh","Gayatri","Heena","Himesh","Hritik","Iriana","Jack","Kalpana","kartik","Khusi","Laura","Laira","Mohan","Monica","Nimesh","Nancy"]

#Image
img=ImageTk.PhotoImage(Image.open("C:\\Users\\nisha\\OneDrive\\Desktop\\Trainings\\PD07\\background.jpg"))
l=Label(image=img)
l.pack()

#Heading Label
app_label=tk.Label(window_main,text="BABY NAMES GENERATOR", font=('calibre',20,'bold'))
app_label.place(x=120,y=30)

#Label
name_label=tk.Label(window_main,text="Baby Name should start from", font=('calibre',12,'bold'))
name_label.place(x=100,y=160)

#Entry
a_var=tk.StringVar()
alphabet_entry=tk.Entry(window_main,textvariable=a_var,font=('calibri',50,'italic'))
alphabet_entry.place(x=360,y=120,width=100)
#to display a element tkinter uses many ways - grids, pack, place (geometry)

def displaynamesFunc():
    for i in babynamesDB:
        if i[0][0]==a_var.get().upper():
            inputtxt.insert(END,(i+" "))

#Display_Button
display_button=tk.Button(window_main,text="Display Baby Name Suggestions",command=displaynamesFunc,font=('calibri',20,'bold'))
display_button.place(x=115,y=280)

#TextInput or TextArea
textarea_var=tk.StringVar()
inputtxt = tk.Text(window_main,font=('calibri',12))
inputtxt.place(x=30,y=370,height=140,width=540)


def clearFunc():
    a_var.set("")
    inputtxt.delete("1.0","end")
    
#Clear Button
clear_button=tk.Button(window_main,text="Clear All",command=clearFunc, font=('calibri',11,'bold'))
clear_button.place(x=400,y=550)

def exitFunc():
    window_main.destroy()
    
#Exit Button
exit_button=tk.Button(window_main,text="Exit",command=exitFunc,font=('calibri',11,'bold'))
exit_button.place(x=490,y=550)




#run the application
window_main.mainloop()
```

```python
window_main=tk.Tk(className='Image change')
window_main.geometry('600x600')

imagedict={"A":"C:\\Users\\nisha\\OneDrive\\Desktop\\Trainings\\PD07\\a.jpg", "B":"C:\\Users\\nisha\\OneDrive\\Desktop\\Trainings\\PD07\\b.jpg"}
#Image

img=ImageTk.PhotoImage(Image.open("C:\\Users\\nisha\\OneDrive\\Desktop\\Trainings\\PD07\\background.jpg"))
l=Label(image=img)
l.pack()

#Heading Label
app_label=tk.Label(window_main,text="BABY NAMES GENERATOR", font=('calibre',20,'bold'))
app_label.place(x=120,y=30)

#Label
name_label=tk.Label(window_main,text="Baby Name should start from", font=('calibre',12,'bold'))
name_label.place(x=100,y=160)

#Entry
a_var=tk.StringVar()
alphabet_entry=tk.Entry(window_main,textvariable=a_var,font=('calibri',50,'italic'))
alphabet_entry.place(x=360,y=120,width=100)
#to display a element tkinter uses many ways - grids, pack, place (geometry)

def displaynamesFunc():
    imgforalpha=ImageTk.PhotoImage(Image.open(imagedict.get(a_var.get().upper())))
    image_label.configure(image=imgforalpha)
    image_label.image=imageforalpha
    for i in babynamesDB:
        if i[0][0]==a_var.get().upper():
            inputtxt.insert(END,(i+" "))

#Display_Button
display_button=tk.Button(window_main,text="Change Image",cursor="clock", bg="blue",fg="white",command=displaynamesFunc,font=('calibri',20,'bold'))
display_button.place(x=115,y=280)

#TextInput or TextArea
textarea_var=tk.StringVar()
inputtxt = tk.Text(window_main,font=('calibri',12))
inputtxt.place(x=30,y=370,height=140,width=300)


#Label
image_label=tk.Label(window_main, borderwidth=10,font=('calibre',12,'bold'))
image_label.place(x=370,y=320,height=200,width=200)

def clearFunc():
    a_var.set("")
    inputtxt.delete("1.0","end")
    
#Clear Button
clear_button=tk.Button(window_main,text="Clear All",command=clearFunc, font=('calibri',11,'bold'))
clear_button.place(x=400,y=550)

def exitFunc():
    window_main.destroy()
    
#Exit Button
exit_button=tk.Button(window_main,text="Exit",command=exitFunc,font=('calibri',11,'bold'))
exit_button.place(x=490,y=550)




#run the application
window_main.mainloop()
```
